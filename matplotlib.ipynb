Matplotlib

Purpose: Data visualization.

Why Use It?

Seeing data visually often reveals patterns and trends better than raw numbers. Matplotlib allows you to create a variety of plots, from simple line graphs to complex 3D visualizations.

Key Features:

Create line plots, bar charts, histograms, scatter plots, etc.
Fully customizable: Titles, labels, colors, styles, and more.
Save plots as images for reports.


```python
#pip install matplotlib
import matplotlib.pyplot as plt
#Simple Plot
x = [1,2,3,4]
y=[10,20,30,40]
plt.plot(x,y)
plt.title("Simple Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#Core Componenets
#Figures and Axes
#Creating a Figure and Axes
fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
#Adding Multiple Axes
fig, axs = plt.subplots(2,3) # 2x2 grid of the subplots
axs[0,0].plot(x,y)
axs[1,1].scatter(x,y)
plt.show()
```

```python
#Plot Types
#Line Plot
plt.plot(x, y, linestyle='--', color='r', marker='o', label='Line')
plt.legend()
plt.show()
```

```python
#Scatter Plot
plt.scatter(x, y, color='green')
plt.show()
```

```python
#Bar Chart
categories = ['A', 'B', 'C']
values = [10, 20, 15]
plt.bar(categories, values, color='orange')
plt.show()
```

```python
#Histogram
data = [1, 1, 2, 2, 3, 4, 5, 6, 7, 8]
plt.hist(data, bins=5, color='green')
plt.show()
```

```python
#Pie Chart
sizes = [25, 35, 20, 20]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
#Box Plot
data = [7, 8, 8, 9, 10, 10, 11, 11, 11, 12]
plt.boxplot(data)
plt.show()
```

```python
#Customization
#Titles and Labels
plt.title("Customized Plot")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
```

```python
#Adding Legends
plt.plot(x, y, label='Line 1')
plt.plot(x, [15, 25, 35, 45], label='Line 2')
plt.legend()
plt.show()
```

```python
#Colours and Styles
plt.plot(x, y, color='purple', linestyle='--', marker='*')
```

```python
#Grids and Annotations
plt.plot(x, y)
plt.grid(True)
plt.annotate("Peak", xy=(3, 25), xytext=(2, 30), 
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.show()
```

```python
#Advanced Visualizations
#Multiple plots
x1 = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
x2 = [1, 2, 3, 4]
y2 = [30, 25, 20, 10]

plt.plot(x1, y1, label="Line 1")
plt.plot(x2, y2, label="Line 2")
plt.legend()
plt.show()
```

```python
#Subplots
fig, axs = plt.subplots(2)
axs[0].plot(x, y)
axs[1].scatter(x, y)
plt.show()
```

```python
#Logrithmic Scales
plt.plot(x, y)
plt.yscale('log')
plt.show()
```

```python
#Error Bars
errors = [1, 2, 1.5, 1]
plt.errorbar(x, y, yerr=errors, fmt='o', ecolor='red')
plt.show()
```

```python
#Interactivity # Adding Buttons
from matplotlib.widgets import Button

def on_click(event):
    print("Button clicked!")

fig, ax = plt.subplots()
button_ax = plt.axes([0.7, 0.01, 0.1, 0.075])
button = Button(button_ax, "Click Me")
button.on_clicked(on_click)
plt.show()
```

```python
#Interactive Pan and Zoom
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
#3D Visualizations

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
z = [5, 15, 25, 35]
ax.scatter(x, y, z)
plt.show()
```

```python
#Animations
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()
x, y = [], []

def animate(i):
    x.append(i)
    y.append(i**2)
    ax.clear()
    ax.plot(x, y)

ani = FuncAnimation(fig, animate, frames=10, repeat=False)
plt.show()
```

```python
#Practical Scenarios
#Data Analysis
#Visualizing trends in datasets using the Pandas and Matplotlib.
import pandas as pd

data = {'Year': [2000, 2005, 2010, 2015, 2020],
        'Population': [2.5, 2.8, 3.2, 3.6, 4.0]}
df = pd.DataFrame(data)
plt.plot(df['Year'], df['Population'])
plt.title("Population Growth Over Years")
plt.show()
```

```python
#Financial Data
stock_prices = [100, 105, 110, 120, 125]
days = [1, 2, 3, 4, 5]
plt.plot(days, stock_prices)
plt.title("Stock Price Over Time")
plt.xlabel("Days")
plt.ylabel("Stock Price")
plt.show()
```

```python
#Scientific Data
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title("Sine Wave")
plt.show()
```

```python
#Machine Learning Visualization

from sklearn.datasets import make_blobs

data, labels = make_blobs(n_samples=100, centers=3, random_state=42)
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis')
plt.title("Cluster Visualization")
plt.show()

```
